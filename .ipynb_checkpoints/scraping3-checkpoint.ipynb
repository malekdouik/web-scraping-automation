```python
def write_infile_configureScraping(url,baliz=[],class_name=[],xpath=[]):
    # write data in a file
    a="["+url+"]"
    with open("test1.ini", "a") as file1:
        file1.write(a)
        file1.write("\n")
        file1.write("url =" + url+'\n')
        file1.write("baliz =" + str(baliz)+'\n') 
        file1.write("class_name =" + str(class_name)+'\n')
        file1.write("xpath =" + str(xpath)+'\n')
        file1.write("\n \n")
```

```python
def write_infile_urls(lien,url=[]):
    with open("test2.ini", "a") as file1:
        # convert list of list into list 
        flat_list = []
        for sublist in url:
            for item in sublist:
                flat_list.append(item)
        
        file1.write("\n")
        file1.write("["+lien+"]")
        file1.write("\n")
        file1.write("recherhce = ")
        file1.write(flat_list[0])
        del flat_list[0]
        file1.write("\n")
        file1.write("url =" + str(flat_list))
        file1.write("\n\n")
```

```python
# read params from config file 
def read_file_function(url):
    ## read file config
    config = ConfigParser()
    config.read('test1.ini') 
    
    ### read entity
    baliz= config[url]['baliz']
    class_name=config[url]['class_name']
    lien = config[url]['url']
    xpath = config[url]['xpath']
  # Convert str to list
    if(len(baliz)!=0 and len(class_name)!=0):
        baliz=ast.literal_eval(baliz) 
        class_name=ast.literal_eval(class_name) 
    if(len(xpath)!=0):
        xpath=ast.literal_eval(xpath)
        
    return lien,baliz,class_name,xpath
```

```python
# FONCTION 1
def scraping_function(lien,tags=[],class_name=[],xpath=[]):    
    url = lien
    article = Article(url)
    article.download()
    article.parse()

#- Bfs4
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page, 'html.parser')

# - Newspaper3K :
    authors=article.authors
    publish_date=article.publish_date
    text=article.text
    top_image=article.top_image
    summary=article.summary

# - Beautifulsoup
    resultat=[]
    
    df1 = pd.DataFrame({'' : []})
    for i in zip(tags, class_name):
    # resultat.append(soup.findAll(tags, {"class": class_name}))
        for each in soup.findAll(tags, class_=class_name):
            resultat.append([each.text.strip()])
            if not(len(resultat) == 0):
                df1 = pd.DataFrame({'result_scraping':resultat[0] })

# - Beautifulsoup + Xpath
    r = requests.get(url)
    doc = html.fromstring(r.content)
    resultat_xpaths=[]

    # convert lower case
    xpath = [x.lower() for x in xpath]
    for i in xpath:
        link = doc.xpath(i)  
        for j in link:
            resultat_xpaths.append(j.text)
            if not(len(resultat_xpaths) == 0):
                df2 = pd.DataFrame({'xpaths':resultat_xpaths })

# - Storage data in dataframe 
    df = pd.DataFrame({
    'Authors':[authors],
    'Publish_date':[publish_date],
    'Text':[text],
    'image':[top_image],
    'Summary':[summary]
    })

    return df,df1,df2
```
